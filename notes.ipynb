## What is an algorithm? 

It is a **set of steps needed to complete a task**. To implement one: 

1 - We need to have a *clear problem statement* 

2 - A *specific order* for the steps 

3 - *Simple and distinct* steps 

To determine the  **efficiency** of an algorithm, its **worst case scenario** should be analyzed using two paradigms:

1 - *Time complexity:* the less it takes to complete the task, the better 

2 - *Space complexity:* the less memory is used, the better

### Binary search

Let's define a searching algorithm: binary search takes a **sorted** data list and returns the position in the list 
or a `None` type if target is not in the list. 

Binary seach is a good basic searching algorithm due to its *sublinear* time complexity. It takes `O(log n)` time.

## 1. What are data structures? 

The collection of values, the format they are stored in, their relationships to one another and 
the operations applied on them. 

Each data structure addresses a particular problem.

### 1.1 Types of insertions

- **Inserting**: Linear runtime. Adds item to an existing index.
- **Appending**: Constant runtime: in the average, has ammortized constant space 
complexity. Adds item to the end of the list. 
- **Extending**: `O(k)`, where `k = number of elements in the list` we are adding.
Adds a list to another list by appending every element to the new list. 

### 1.2 Deletion 

Has an upper bound of `O(n)` (linear runtime).

## 2. Types of data structures 

### 2.1 Arrays (native to most programming languages) 

A **contiguous** data structure, i.e it is stored in blocks in memory that are beside each other with no gaps. 
For this reason, accessing data in an array/Python list is a constant time operation.

Arrays are good for reading, but not that much for inserting and deleting.

### 2.2 Linked Lists

Linear data structure where each element is contained in a separate object called a **node**.A node consists of the actual data 
and a reference to the next node (or both the next and previous node in case of doubly linked lists). The first node in the list 
is called the **head** and last one the **tail**. 

*The list only maintains a reference to the head*.

Better for inserting/deleting than array[<sup>[1]</sup>](#references).
